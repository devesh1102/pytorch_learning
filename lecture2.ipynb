```python
import torch
import torchvision
from torchvision import transforms, datasets
import matplotlib.pyplot as plt
```

here we use dataset from torch vision we see

```python
train = datasets.MNIST('', train=True, download=True,transform=transforms.Compose([transforms.ToTensor()]))

test = datasets.MNIST('', train=False, download=True,
                       transform=transforms.Compose([
                           transforms.ToTensor()
                       ]))
```

we can use the buildin funnction on pytorch to convert the data into batches and then iterate over the batches

```python
trainset = torch.utils.data.DataLoader(train, batch_size=10, shuffle=True)
testset = torch.utils.data.DataLoader(test, batch_size=10, shuffle=False)
```

here data has the features of all the batches. we have 10 batch each entry inn batch is a datapoint. each datapoint ie image in a tensor of 1* 28* 28
