```python
import torch
import torchvision
from torchvision import transforms, datasets
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
```

downloading the. dataset and transforming it into a tensor

```python
train = datasets.MNIST('', train=True, download=True,transform=transforms.Compose([transforms.ToTensor()]))

test = datasets.MNIST('', train=False, download=True,
                       transform=transforms.Compose([
                           transforms.ToTensor()
                       ]))
```

converting the dataset into batches of size 10

```python
batch = 10
trainset = torch.utils.data.DataLoader(train, batch_size=batch , shuffle=True)
testset = torch.utils.data.DataLoader(test, batch_size=batch , shuffle=False)
layer1 = 784
layer2 = 64
layer3 = 64
layer4 = 64
```

```python
class network(nn.Module):
    def __init__(self):
        super().__init__()
#         for initialization
        self.l1 = nn.Linear(layer1,layer2)
        self.l2 = nn.Linear(layer2,layer3)
        self.l3 = nn.Linear(layer3,layer4)
        self.l4 = nn.Linear(layer4,10)
    def forward(self,x):
        x= F.relu(self.l1(x))
        x= F.relu(self.l2(x))
        x= F.relu(self.l3(x))
        x= self.l4(x)
        return F.log_softmax(x,dim = 1)
        
net = network()
print(net)
#         self.l2 = nn.Linear(28*28,64)/
        
            
```

```python
optimizer = optim.Adam(net.parameters(),lr=0.001)
epochs=3
for epoch in range(epochs):
    for data in trainset: 
        x,y= data
        net.zero_grad()
#         print(data.shape,data.view(-1,28*28).shape)
#         break
        output = net(x.view(-1,28*28) )
        loss = F.nll_loss(output,y)
        loss.backward()
        optimizer.step()
    print(loss)
```

testing

```python
total = 0
wrong = 0
with torch.no_grad():
    for data in testset:
        x,y = data
        output = net(x.view(-1,28*28))
#         print(output.shape)
#         print(output[0],y[0])
        for i in range (batch):
            if torch.argmax(output[i])  != y[i]:
                plt.imshow(x[i].view(28,28))
                plt.show()
                print(y[i],torch.argmax(output[i]) )
                wrong = wrong +1
            total = total +1
print(total,wrong)
#         break
        
```
